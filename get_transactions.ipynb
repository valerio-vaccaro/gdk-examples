# Bitcoin
## Mainnet multisig

```python
import greenaddress
try:
    greenaddress.init({})
except:
    pass

gdkMnemonic = 'order earth buddy render ocean produce bacon orchard congress law illness goat'
if not greenaddress.validate_mnemonic(gdkMnemonic):
    raise Exception("Invalid mnemonic.")
        
s = greenaddress.Session({'name': 'mainnet', 'log_level': 'info'})

s.login_user({}, {'mnemonic': gdkMnemonic}).resolve()
received_txs = []
index = 0
amount = 1
while(True):
    transactions = s.get_transactions({'subaccount': 0,'first': index, 'count':amount}).resolve()
    for transaction in transactions['transactions']:
        received_txs.append(transaction)
    if len(transactions['transactions']) < amount:
        break
    index = index + 1
received_txs
```
